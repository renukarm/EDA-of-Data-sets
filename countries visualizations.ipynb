```python
import pandas as pd
import numpy as np
df=pd.read_csv(r'C:\Users\mssre\Desktop\datas\world report.csv',"\t")
df.head()
```

```python
#data cleaning
df.dropna(inplace=True)
print(df.isna().sum())
print(df.duplicated().sum())
```

```python
#data visualization, classifying off of GDP
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.scatterplot(y="Logged GDP per capita",x="Country name",data=df,hue="Logged GDP per capita")
plt.xticks(rotation=90,size=4)
plt.legend(bbox_to_anchor=(1.05,1),loc=2)
plt.show()
```

```python
#data visualization, top 10 of everything
def top10(value):
    return df[value].sort_values(ascending=False).iloc[0:10]
cols=df.columns
for name in cols:
    print("The tabele is of",name,", here it is:",top10(name))
```

```python
bols=df.columns
new_df=df.drop(bols[14:21],axis=1)
new=new_df.drop(["Standard error of ladder score","upperwhisker","lowerwhisker"],axis=1)
happs=new.columns
print(happs)
print(new.shape)
```

```python
#all visualizations in one code
happs=new.columns
gud=[]
for j in happs:
    gud.append(j)
print(gud)
guds=gud[3:12]
print(guds)

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

def vas(rix):
    l=new[rix].mean()
    return new[new[rix]>=l].sort_values(rix,ascending=False)
for h in guds:
    sns.lineplot(x="Country name",y=h,data=vas(h))
    plt.title(h)
    plt.xticks(rotation=90,size=5)
    plt.show()
    sns.lineplot(x="Regional indicator",y=h,data=vas(h))
    plt.title(h)
    plt.xticks(rotation=90)
    plt.show()
    print("Details of top happy mean places according to **",h,"** are: ",vas(h))
```
